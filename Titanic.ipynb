```python
from sklearn.model_selection import train_test_split
```

```python
remove=['Fare','2urvived','Embarked']
X=df.drop(remove, axis=1)
y=df['2urvived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
pred=dtree.predict(X_test)
```

```python
pred
```

```python
value=pd.DataFrame({'survivors': pred})
value.set_index(pd.Index(range(len(pred))), inplace=True)
value.index.name = 'passenger_id'
value.head()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
print('\n')
print(confusion_matrix(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc_pred=rfc.predict(X_test)
```

```python
rfc_pred
```

```python
result=pd.DataFrame({'survivors': rfc_pred})
result.set_index(pd.Index(range(len(rfc_pred))), inplace=True)
result.index.name = 'passenger_id'
result.head()
```

```python
print(classification_report(y_test,rfc_pred))
print('\n')
print(confusion_matrix(y_test,rfc_pred))
```
